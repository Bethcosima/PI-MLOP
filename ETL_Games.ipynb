```python
import pandas as pd 
import numpy as np
import sys
import json
import ast
import locale
```

```python
#abrimos el archivo df
df=[]
with open(r'output_steam_games.json', encoding='utf-8-sig') as f:
    for linea in f:
        if ' "id":NaN' in linea:
            continue
        linea = linea.replace('false', 'False')
        linea = linea.replace('true','True')
        linea = linea.replace('NaN', 'None')
        df.append(eval(linea))

df_games= pd.DataFrame(df)
```

```python
df_games.head()
```

```python
#verificamos los duplicados
df_games[df_games['id'].duplicated(keep=False)]
```

```python
#se eliminan los duplicados
df_games.drop_duplicates(subset='id', keep='first', inplace=True)
```

```python
#eliminamos la primera fila que contiene valores NaN o/y None el subset 
#eliminará la fila si todos los valores son Nan o None
#df_games=df_games.dropna(subset=df_games.columns, how='all').reset_index(drop=True)
```

```python
#convertimos las fechas en formato date
df_games['release_date']= pd.to_datetime(df_games['release_date'], format='%Y-%m-%d', errors='coerce')
```

```python
# dejamos unicamente el año en release_date

df_games['release_date'] = df_games['release_date'].dt.year
```

```python
#eliminamos los valores nulos de la columna release_date
df_games['release_date'] =df_games['release_date'].dropna()
```

```python
#eliminamos todos los valores nulos del df
df_games= df_games.dropna()
```

```python
df_games.head()
```

```python
#Funcion para normalizar los precios, convertimos los free a 0

def ceros(c):
    if isinstance(c,str):
        return 0
    return c
    
#aplicamos la funcion a la columna 'price' del DF

df_games['price']= df_games['price'].apply(ceros)
```

```python
#convertimos a flotante
df_games['price'].astype(float)
```

```python
#cambiamos el nombre de id por item_id
df_games.rename(columns={'id':'item_id'}, inplace=True)
```

```python
#eliminamos las filas con valores nulos en item_id
f= df_games[df_games.item_id.isnull()].index
df_games.drop(f, inplace=True)
df_games.reset_index(inplace=True, drop=True)
```

```python
f= df_games[df_games.genres.isnull()].index
df_games.drop(f,inplace=True)
df_games.reset_index(inplace=True, drop=True)
```

```python
#creamos una columna nueva para el año de lanzamiento que no contenga .0
df_games['año_lanzamiento']= df_games['release_date'].apply(lambda x: str(x)[:4] if str(x)[:4].isnumeric() else 0)
```

```python
#eliminamos la columna release_date
df_games.drop(['release_date'], axis=1, inplace=True)
```

```python
#comprobamos que no se encuentren valores faltantes
df_games.isna().sum()
```

```python
#eliminamos la columna de app_name
df_games.drop(['app_name'],axis=1, inplace=True)
```

```python
#guaramos el DF en un csv
df_games.to_csv(r'output_games.csv', index=False, errors='replace')
```
