```python
import pandas as pd 
import json
import numpy as np
import ast 
import locale
```

```python

#la ruta del archivo

#abrimos el archivo para procesarlo
with open("australian_users_items.json", 'r', encoding="utf-8") as f:
    data_items= f.readlines()

items_list= [eval(line.strip())for line in data_items]

#convertimos a un dataframe
df_items =pd.DataFrame(items_list)
```

```python
df_items.head()
```

```python
#quitamos los corchetes del df 
df_items= df_items.explode('items')

#creamos listas para guardar los datos qeu desanidaremos
item_id =         []
item_name =       []
playtime_forever= []
playtime_2weeks=  []

#recorremos las filas del DF para poder almacenar en las listas su contenido
for index, row in df_items.iterrows():
    #obtenemos el valor de la columna 'items' para la fila actual
    elem= row['items']

    #verificamos el valor del diccionario
    if isinstance(elem, dict):

        #se extraen los valores especificos en cada diccionario para agregarlo al df
        item_id.append(elem.get('item_id', None)) #agregamos el elemento correspondiente
        item_name.append(elem.get('item_name', None))
        playtime_forever.append(elem.get('playtime_forever', None))
        playtime_2weeks.append(elem.get('playtime_2weeks', None))

    else:
        #si no encontramos el valor en el diccionario agregamos None a la lista
        item_id.append(None)
        item_name.append(None)
        playtime_forever.append(None)
        playtime_2weeks.append(None)

#agregamos las listas c√≥mo columnas nuevas en el df

df_items['Item id']= item_id
df_items['Item name']= item_name
df_items['Playtime Forever']= playtime_forever
df_items['Playtime 2weeks']= playtime_2weeks
```

```python
df_items.drop('items', axis=1, inplace=True)
```
